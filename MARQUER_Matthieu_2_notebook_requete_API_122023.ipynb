```python
# Version revisitée
from stackapi import StackAPI
import pandas as pd

# Instance de StackAPI
SITE = StackAPI('stackoverflow')

# Requête de récuperation des questions
questions = SITE.fetch('questions', fromdate=1457136000, todate=1757222400, # 1457222400
                       min=51, tagged='python', sort='votes', pagesize=10, max_pages=1)['items'] # pagesize=50

questions_list = []

for question in questions:
    quest = {
        'date': pd.to_datetime(question['creation_date'], unit='s'),  # convertir le timestamp en format de date
        'title': question['title'],
        'tags': question['tags'],
        'score': question['score']
    }
    
    questions_list.append(quest)
    
data = pd.DataFrame(questions_list)
data
```
